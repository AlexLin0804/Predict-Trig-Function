```python

import numpy as np
import matplotlib.pyplot as plt
x_train = np.arange(-6*np.pi, 6*np.pi, 0.1)
y_train = np.sin(x_train)
x_test = x_train
y_test = y_train
x_extend = np.arange(-6*np.pi, 6*np.pi, 0.1)
y_extend = np.sin(x_extend)

plt.figure(figsize=[12, 4])
ax1 = plt.subplot(1, 2, 1)
ax2 = plt.subplot(1, 2, 2)

plt.sca(ax1)
plt.plot(x_train, y_train, marker='.', color='r', linestyle='-')
plt.grid()

plt.sca(ax2)
plt.plot(x_extend, y_extend, marker='.', color='b', linestyle='-')
plt.grid()
```

```python


import numpy as np
import keras
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
from keras.models import Model

# config
eta  = 0.025        # learning rate 
num_epochs = 2000
b_size = 126   # batch size



# define model architecture
model = Sequential()
model.add(Dense(40, input_dim=1, activation='tanh'))   # add hidden layer
model.add(Dense(20,  activation='tanh'))   # add hidden layer
model.add(Dense(5,  activation='tanh'))   # add hidden layer
model.add(Dense(1, activation='linear'))  # add output layer 


# optimizer
opt = keras.optimizers.Adam(learning_rate=eta)


# build model 
model.compile(loss='mean_squared_error',
             optimizer=tf.keras.optimizers.Adam(0.001),
             metrics=[tf.keras.metrics.MeanAbsoluteError()])

# train & evaluate 
model.fit(x_train,y_train, epochs=num_epochs, batch_size=b_size)
score = model.evaluate(x_test, y_test, batch_size=b_size)

# retrieve results 
x_test = x_train

final = Model(inputs=model.input, outputs=model.layers[3].output)
final_output = final.predict(x_test)

final_output
```
